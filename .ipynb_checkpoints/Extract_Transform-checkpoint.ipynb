```python
import pandas as pd
import math
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
# Read the data
gender_df = pd.read_csv('data/gender_submission.csv')
test_df = pd.read_csv('data/test.csv')
train_df = pd.read_csv('data/test.csv')
```

```python
# storing passenger names and id
passenger_names = train_df[['PassengerId','Name']]
passenger_names = passenger_names.set_index(passenger_names['PassengerId'])
passenger_names = passenger_names.drop('PassengerId', axis=1)

passenger_names_train_df = test_df[['PassengerId','Name']]
passenger_names_train_df = passenger_names_train_df.set_index(passenger_names_train_df['PassengerId'])
passenger_names_train_df = passenger_names_train_df.drop('PassengerId', axis=1)
```

```python
# removing passenger names as OHC would assign each one a random value, might attrbute to overfitting.
train_df = train_df.drop(['Name'], axis=1)
test_df = test_df.drop(['Name'], axis=1)
```

```python
# use an encoder to transform Sex, Cabin, and Embarked to a real number.
for row, val in train_df.iterrows():
    if val['Sex'] == 'male':
        train_df.Sex[row] = 0
    if val['Sex'] == 'female':
        train_df.Sex[row] = 1


for row, val in test_df.iterrows():
    if val['Sex'] == 'male':
        test_df.Sex[row] = 0
    if val['Sex'] == 'female':
        test_df.Sex[row] = 1
```

```python
# Encoding Cabin column.
ohc = OneHotEncoder()
ohc.fit(train_df[['Cabin']])
encoded_cabin_categories = ohc.categories_
x = ohc.transform(train_df[['Cabin']]).toarray()
encoded_cabin = pd.DataFrame(x)
encoded_cabin.head()

ohc = OneHotEncoder()
ohc.fit(test_df[['Cabin']])
encoded_cabin_categories = ohc.categories_
x = ohc.transform(test_df[['Cabin']]).toarray()
encoded_cabin_test_df = pd.DataFrame(x)
encoded_cabin_test_df.head()
```

```python
# Encoding Embarked column.
ohc = OneHotEncoder()
ohc.fit(train_df[['Embarked']])
encoded_embarked_categories = ohc.categories_
x = ohc.transform(train_df[['Embarked']]).toarray()
encoded_embarked = pd.DataFrame(x)
encoded_embarked.head()

ohc = OneHotEncoder()
ohc.fit(test_df[['Embarked']])
encoded_embarked_categories = ohc.categories_
x = ohc.transform(test_df[['Embarked']]).toarray()
encoded_embarked_test_df = pd.DataFrame(x)
encoded_embarked_test_df.head()
```

```python
# replace Embarked columns with encoded columns
train_df = train_df.join(encoded_embarked)
train_df = train_df.drop('Embarked', axis=1)

test_df = test_df.join(encoded_embarked_test_df)
test_df = test_df.drop('Embarked', axis=1)
```

```python
# encoding Cabin column.
ohc = OneHotEncoder()
ohc.fit(train_df[['Cabin']])
encoded_cabin_categories = ohc.categories_
x = ohc.transform(train_df[['Cabin']]).toarray()
encoded_cabin = pd.DataFrame(x)

ohc = OneHotEncoder()
ohc.fit(test_df[['Cabin']])
encoded_cabin_categories_test_df = ohc.categories_
x = ohc.transform(test_df[['Cabin']]).toarray()
encoded_cabin_test_df = pd.DataFrame(x)
encoded_cabin_test_df.head()
```
